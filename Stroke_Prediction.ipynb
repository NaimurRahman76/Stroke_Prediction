```python
from sklearn.model_selection import train_test_split
train_data, test_data = train_test_split( data , test_size =0.1,random_state=0,stratify=data['stroke'])
```

```python
x_train=train_data;
y_train=train_data['stroke']
x_train=x_train[x_train.columns.drop('stroke')]


x_test=test_data;
y_test=test_data['stroke']
x_test=x_test[x_test.columns.drop('stroke')]
```
